```python
# Official packages
import os
import copy

# Third-party packages
import numpy as np
import matplotlib.pyplot as plt

# In-house packages
from BrukerSpSpEpi import BrukerSpSpEpiExp
```

```python
exp_dir = "../20230305_163320_AgroseCylinder2_1_1"
exp_nbr = 21
        
spsp_epi = BrukerSpSpEpiExp(exp_dir, exp_nbr)
```

```python
plt.plot((spsp_epi.fid['raw']))
```

```python
plt.plot(np.real(spsp_epi.fid['deserialized']))
plt.plot(np.imag(spsp_epi.fid['deserialized']))
```

```python
np.shape(spsp_epi.fid['deserialized'])
```

```python
reformatted_fid_2d = np.reshape(spsp_epi.fid['deserialized'], (spsp_epi._exp_data_dim_dict['dim_k_raw_ph'],-1))

plt.imshow(np.real(reformatted_fid_2d), cmap="seismic", vmin=-1*2**20, vmax=2**20)
```

```python
def mirror_odd_lines(fid_2d):
    mirrored_fid_2d = copy.deepcopy(fid_2d)
    for line_nbr, line in enumerate(mirrored_fid_2d):
        if (line_nbr % 2 ):
             mirrored_fid_2d[line_nbr] = line[::-1,...]
    return mirrored_fid_2d

mirrored_fid_2d = mirror_odd_lines(reformatted_fid_2d)

plt.imshow(np.real(mirrored_fid_2d), cmap="seismic", vmin=-1*2**20, vmax=2**20)
```

```python
zerofilled_fid_2d = np.pad(mirrored_fid_2d, pad_width=((40, 0), (0, 0)))

plt.imshow(np.real(zerofilled_fid_2d), cmap="seismic", vmin=-1*2**20, vmax=2**20)
```

```python
fft_fid_2d = (np.fft.fft2(zerofilled_fid_2d))

plt.imshow(np.real(fft_fid_2d), cmap="seismic", vmin=-1*2**31, vmax=2**31)
```

```python
fftShift_fid_2d = np.fft.fftshift(fft_fid_2d)


plt.imshow(np.real(fftShift_fid_2d), cmap="seismic", vmin=-1*2**31, vmax=2**31)
```
